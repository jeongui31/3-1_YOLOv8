```python
from PIL import Image, ImageOps
import os

def add_padding_to_images(input_folder, output_folder, padding_size):
    # 입력 폴더와 출력 폴더가 존재하는지 확인하고, 없으면 생성합니다.
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 입력 폴더의 모든 파일을 반복합니다.
    for filename in os.listdir(input_folder):
        # 파일이 이미지인지 확인합니다.
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            # 이미지 파일을 엽니다.
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path)
            
            # 패딩을 추가합니다.
            padded_img = ImageOps.expand(img, border=padding_size, fill=128)  # 흰색 패딩
            
            # 패딩이 추가된 이미지를 저장합니다.
            output_path = os.path.join(output_folder, filename)
            padded_img.save(output_path)
            print(f'Processed {filename}')

input_folder = './images/test'  # 입력 폴더 경로
output_folder = './padding'  # 출력 폴더 경로
padding_size = 6  # 패딩 크기 (픽셀 단위, 위아래좌우 모두 적용)

add_padding_to_images(input_folder, output_folder, padding_size)
```
